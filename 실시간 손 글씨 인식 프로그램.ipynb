### 실시간 손글씨 인식 프로그램
>- split()
>- hsplit(), vsplit()

```python
import cv2
import numpy as np
img = cv2.imread('digits.png', 0)
```

```python
print(img.shape)
```

```python
cv2.imshow('numbers',img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
cells = [np.hsplit(row, 100)for row in np.vsplit(img, 50)]
```

```python
len(cells), len(cells[0])
```

```python
x = np.array(cells)
x.shape
# 100개씩 총 50개의 행
# 5000행 400열로 만들어야 함
```

```python
X = x.reshape(-1, 400).astype(np.float32)
# 한개의 데이터는 한개의 차원에 할당되기 때문에 변환 필요
```

```python
X.shape
```

```python
y = np.repeat(np.arange(10),500)
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split

# 알아서 셔플 되니까 걱정하지말고 해도 됨
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size = 0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
model.fit(train_X, train_y)
# 학습
```

```python
model.score(test_X, test_y)
# 실제 현업에 적용하면 저렇게 나올까?
```

```python
# 모델을 파일에 저장해 놓으면 사용하기 편리함
# 피클로 저장할 때는 바이너리로 저장해야함

import pickle

with open('number.model', 'wb') as f:
    pickle.dump(model,f)
```

```python
with open('number.model', 'rb') as f:
    model2 = pickle.load(f)
```

```python
model2
```

```python
model2.score(test_X, test_y)
```
